```python
from datasets import load_from_disk
import json
from transformers import AutoTokenizer
```

```python
processed = load_from_disk('../data/processed')
print(processed)

# Load label config
with open('../data/processed/label_config.json', 'r') as f:
    config = json.load(f)

# Check sample
sample = processed['train'][0]
tokenizer = AutoTokenizer.from_pretrained('vinai/phobert-base-v2')

print("\nSample tokens and labels:")
for i in range(20):
    token = tokenizer.convert_ids_to_tokens([sample['input_ids'][i]])[0]
    label = config['id_to_label'].get(str(sample['labels'][i]), 'IGNORE')
    print(f"{token:20} → {label}")
```
